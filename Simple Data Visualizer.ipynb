```python
# Simple Data Visualization using Matplotlib

import matplotlib.pyplot as plt

# Sample data
students = ["Asha", "Rahul", "Meena", "Vivek", "Riya"]
marks = [78, 85, 69, 92, 74]

# Create a bar chart
plt.bar(students, marks)
plt.title("Students' Marks Comparison")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

# Create a pie chart
plt.pie(marks, labels=students, autopct='%1.1f%%', startangle=90)
plt.title("Percentage of Marks by Students")
plt.show()
```
